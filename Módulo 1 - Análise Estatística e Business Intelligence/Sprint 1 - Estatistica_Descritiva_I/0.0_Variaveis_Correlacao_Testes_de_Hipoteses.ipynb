1.   Você quer medir a força e a direção da relação entre duas variáveis? Use **Coeficientes de Correlação**.

### Cheat Sheet
* Calcular correlação: `df['var1'].corr(df['var2'])`
* Plotar um gráfico de dispersão com uma linha de regressão: `sns.regplot(x='var1', y='var2', data=df)`
* Realizar um teste t: `stats.ttest_ind(grupo1, grupo2)`
* Realizar um teste de Mann-Whitney: `stats.mannwhitneyu(grupo1, grupo2)`
* Realizar uma ANOVA: `sm.stats.anova_lm(ols('var ~ C(fator)', data=df).fit(), typ=2)`

### Glossário
* Correlação: Uma medida da força e direção da relação entre duas variáveis. Varia de -1 (correlação negativa perfeita) a 1 (correlação positiva perfeita).
* Teste T: Um teste estatístico usado para comparar as médias de dois grupos independentes. Ele retorna um p-valor, que é usado para determinar se a diferença é estatisticamente significativa.
* Teste de Mann-Whitney: Um teste não paramétrico usado para comparar os ranks de dois grupos independentes. É usado quando os dados não são normalmente distribuídos ou quando os tamanhos das amostras são pequenos.
* ANOVA (Análise de Variância): Um teste estatístico usado para comparar as médias de três ou mais grupos independentes. Ele retorna um p-valor, que é usado para determinar se as diferenças são estatisticamente significativas.
* p-valor: A probabilidade de obter os dados observados (ou dados mais extremos) se a hipótese nula for verdadeira. Via de regra, se o p-valor for menor que 0.05, rejeitamos a hipótese nula e concluímos que a diferença observada é estatisticamente significativa.
* Hipótese Nula: Uma declaração de que não há efeito ou diferença. É a hipótese que tentamos rejeitar em um teste estatístico.
* Linha de Regressão: Uma linha que melhor se ajusta aos pontos de dados em um gráfico de dispersão. É usada para representar a relação entre duas variáveis.
* Gráfico de Dispersão: Um gráfico no qual os valores de duas variáveis são plotados ao longo de dois eixos. É usado para observar e mostrar relações entre duas variáveis numéricas.
* Distribuição Normal: Uma distribuição de probabilidade que é simétrica em relação à média, mostrando que os dados próximos à média são mais frequentes em ocorrência do que os dados distantes da média.
* Teste Não Paramétrico: Um teste estatístico que não assume que os dados seguem uma distribuição particular.
* Tamanho da Amostra: O número de observações em uma amostra.
* Grupos Independentes: Grupos que não estão relacionados entre si de qualquer maneira.
* Fator: Uma variável categórica usada em um experimento ou estudo observacional.
* Significância Estatística: A probabilidade de que um resultado ou relação seja causado por algo além do mero acaso. O teste de hipóteses estatísticas é tradicionalmente empregado para determinar se um resultado é estatisticamente significativo ou não.
* Variável: Qualquer característica, número ou quantidade que pode ser medida ou contada. Elas são chamadas de variáveis porque a informação representada pode mudar (variar) e elas podem assumir muitos valores.
* Array: Uma estrutura de dados que pode conter um número fixo de itens do mesmo tipo. Os itens são armazenados em locais de memória contíguos.
* Módulo: Um arquivo contendo definições e declarações Python. O nome do arquivo é o nome do módulo com o sufixo .py adicionado.
* Função: Um bloco de código organizado e reutilizável que é usado para realizar uma única ação relacionada. As funções proporcionam uma melhor modularidade para sua aplicação e um alto grau de reutilização de código.
* Método: Uma função que está associada a um objeto. Em Python, um método não é único para instâncias de classe. Qualquer tipo de objeto pode ter métodos.



### Coeficiente de Correlação

<br>

**Visão Geral** <br>
Imagine que você está numa festa de dança. A correlação é como uma dança entre duas variáveis. Quando uma variável se move, a outra se move com ela, contra ela, ou não se importa com ela. Se elas se movem juntas na mesma direção, elas têm uma correlação positiva. Se elas se movem em direções opostas, elas têm uma correlação negativa. Se uma variável dança ao seu próprio ritmo sem qualquer consideração pela outra, elas não têm correlação.

<br>

**Casos de Uso e Exemplos de Código** <br>
Vamos supor que você é um professor e quer ver se há uma correlação entre o número de horas que seus alunos estudam e suas notas nos exames.

```python
import pandas as pd
import seaborn as sns

df = pd.DataFrame({
    'Horas_Estudadas': [2, 4, 6, 8, 10],
    'Notas_Exames': [50, 60, 70, 80, 90]
})

correlacao = df['Horas_Estudadas'].corr(df['Notas_Exames'])
sns.regplot(x='Horas_Estudadas', y='Notas_Exames', data=df)
```

Nesse bloco de código, primeiro criamos um DataFrame com as horas estudadas e as notas correspondentes nos exames. Em seguida, calculamos a correlação entre essas duas variáveis usando a função `corr()`. Finalmente, plotamos um gráfico de dispersão com uma linha de regressão para visualizar a correlação.

O valor da correlação será entre -1 e 1.Se o valor estiver próximo de 1, significa que há uma forte correlação positiva (ou seja, quando as horas estudadas aumentam, as notas dos exames também aumentam). Se o valor estiver próximo de -1, significa que há uma forte correlação negativa (ou seja, quando as horas estudadas aumentam, as notas dos exames diminuem). Se o valor estiver próximo de 0, significa que não há correlação.

<br>

**Memorizando** <br>
Lembre-se da metáfora da festa de dança. Correlação positiva é como uma dupla de dança bem coordenada se movendo em sincronia. Correlação negativa é como uma dança rebelde onde um dançarino se move na direção oposta. Não correlação é como dois dançarinos fazendo sua própria coisa, alheios um ao outro.

### Teste T

<br>

**Visão Geral** <br>
Imagine que você é um juiz em um concurso de culinária. Você tem dois cozinheiros, cada um afirmando que seu bolo é o melhor. Você não pode simplesmente aceitar a palavra deles, você precisa provar os bolos. Um teste t é como provar os bolos para ver se há uma diferença significativa no sabor. Se houver, você pode declarar um vencedor. Se não houver, é um empate.

<br>

**Casos de Uso e Exemplos de Código** <br>
Vamos supor que você é um cientista e desenvolveu um novo fertilizante. Você quer saber se ele aumenta significativamente o crescimento das plantas em comparação com o fertilizante antigo.

```python
import numpy as np
from scipy import stats

fertilizante_antigo = np.array([30, 32, 31, 29, 30])
fertilizante_novo = np.array([33, 35, 32, 34, 36])

t_stat, p_val = stats.ttest_ind(fertilizante_antigo, fertilizante_novo)
print(f'p-valor: {p_val}')
```

Nesse bloco de código, primeiro criamos dois arrays com as medidas de crescimento das plantas para os fertilizantes antigo e novo. Em seguida, realizamos um teste t usando a função ttest_ind() do módulo scipy.stats. Finalmente, imprimimos o p-valor. Se o p-valor for menor que 0.05, podemos concluir que o novo fertilizante aumenta significativamente o crescimento das plantas.

O p-valor nos diz a probabilidade de obter os dados observados (ou dados mais extremos) se a hipótese nula for verdadeira. Nesse caso, a hipótese nula é que os dois fertilizantes têm o mesmo efeito no crescimento das plantas. Se o p-valor for menor que 0.05, rejeitamos a hipótese nula e concluímos que o novo fertilizante aumenta significativamente o crescimento das plantas.

<br>

**É possível ter um t_stat alto e um p-valor alto (maior que 0.05)? Caso positivo, como esse resultado seria interpretado?** <br>

Sim, é possível ter um valor alto para a estatística t (t_stat) e, ao mesmo tempo, um p-valor alto (maior que 0.05). Essa combinação de resultados pode ser interpretada da seguinte maneira:

- Valor alto para a estatística t (t_stat): A estatística t mede a magnitude da diferença entre as médias das duas amostras em relação à variabilidade dentro das amostras. Um valor alto para t_stat indica que as médias das amostras são bastante diferentes em termos absolutos.
- P-valor alto (maior que 0.05): O p-valor representa a probabilidade de obter os dados observados (ou dados mais extremos) se a hipótese nula for verdadeira. Nesse contexto, a hipótese nula geralmente afirma que não há diferença significativa entre as médias das duas amostras.

A interpretação desse resultado pode ser a seguinte:

Apesar de haver uma diferença significativa entre as médias das duas amostras (indicada pelo alto valor de t_stat), o p-valor alto sugere que essa diferença pode ser devida ao acaso ou a variações naturais nos dados. Em outras palavras, não há evidências estatísticas suficientes para rejeitar a hipótese nula de que não há uma diferença verdadeira entre os dois grupos. Portanto, não podemos concluir com confiança que o novo fertilizante tem um efeito significativamente maior no crescimento das plantas do que o fertilizante antigo.

Essa situação pode ocorrer em casos em que as amostras são pequenas ou a variabilidade dentro das amostras é alta. Quanto menor o tamanho da amostra ou maior a variabilidade, mais difícil será obter um p-valor pequeno, mesmo que as médias das amostras sejam diferentes.

Em resumo, um t_stat alto com um p-valor alto indica que, apesar de haver uma diferença considerável entre as médias das amostras, não há evidências estatísticas suficientes para afirmar que essa diferença é significativa e não é simplesmente devida ao acaso ou à variabilidade natural dos dados.

<br>

**Memorizando**<br>
Lembre-se da metáfora do concurso de culinária. Um teste t é como provar dois bolos para ver se há uma diferença significativa no sabor. Se o p-valor for menor que 0.05, é como declarar um vencedor no concurso de culinária.

###Teste de Mann-Whitney

<br>

**Visão Geral**<br>
O teste de Mann-Whitney é uma ferramenta estatística utilizada para comparar duas amostras independentes e determinar se existe uma diferença significativa entre elas em termos de suas classificações ou pontuações. Ele é aplicado quando não podemos assumir que os dados seguem uma distribuição normal e desejamos analisar a ordem ou ranking dos valores, em vez de seus valores absolutos.

<br>

**Casos de Uso e Exemplos de Código**<br>
Imagine que você é um empreendedor que possui uma plataforma online para venda de produtos eletrônicos, e você está interessado em avaliar a satisfação dos clientes com dois modelos diferentes de smartphones que você oferece: Modelo X e Modelo Y.

Você coletou as classificações dos clientes em uma escala de 1 a 5 (sendo 1 muito insatisfeito e 5 muito satisfeito) após suas compras. Agora, você deseja determinar se existe uma diferença significativa na satisfação dos clientes entre o Modelo X e o Modelo Y.

```python
#Primeiro, organize os dados em dois conjuntos, um para as classificações do Modelo X e outro para as classificações do Modelo Y:

modelo_X = [1, 2, 3, 4, 1, 4, 2, 3, 5, 4]
modelo_Y = [3, 4, 4, 3, 5, 3, 4, 3, 5, 5]
```

```python
# Em seguida, utilize o teste de Mann-Whitney usando a biblioteca scipy.stats:

import scipy.stats as stats

u_stat, p_val = stats.mannwhitneyu(modelo_X, modelo_Y)
print(f'p-valor: {p_val}')
```

No resultado, o p-valor obtido foi de aproximadamente 0.10. Se adotarmos um nível de significância de 0.05, podemos interpretar que **não há** diferença significativa na satisfação dos clientes entre o Modelo X e o Modelo Y, pois o p-valor é maior do que 0.05. Isso significa que as classificações atribuídas pelos clientes para os dois modelos **não são** estatisticamente diferentes.

Neste exemplo, organizamos as classificações dos clientes em dois conjuntos, representando os dois modelos de smartphones que queremos comparar. Em seguida, aplicamos o teste de Mann-Whitney usando a função mannwhitneyu() do módulo scipy.stats para calcular o p-valor. O p-valor é comparado com um nível de significância (geralmente 0.05) para determinar se há uma diferença estatisticamente significativa entre as classificações dos dois modelos.

<br>

**Memorizando**<br>
Para lembrar o conceito do teste de Mann-Whitney, pense em comparar as classificações de satisfação de clientes entre dois modelos de smartphones, como Modelo X e Modelo Y. O objetivo é determinar se há uma diferença significativa na satisfação dos clientes entre os dois modelos, ou se um deles é consistentemente classificado como mais satisfatório pelos clientes em relação ao outro.

### Análise de Variância (ANOVA)

<br>

**Visão Geral**<br>
Imagine que você é um juiz em um show de talentos com três concorrentes. Cada concorrente realiza um ato diferente, e você quer saber se há uma diferença significativa no aplauso do público para cada ato. ANOVA é como medir o aplauso para cada ato e compará-los para ver se um ato recebe significativamente mais aplausos que os outros.

<br>

**Casos de Uso e Exemplos de Código**<br>
Vamos supor que você é um empresário e quer saber se há uma diferença significativa nas vendas de três produtos diferentes.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

df = pd.DataFrame({
    'Vendas': [100, 120, 110, 150, 200, 180, 220, 80, 70],
    'Produto': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C']
})
```

```python
# Criação do modelo ANOVA:
model = ols('Vendas ~ C(Produto)', data=df).fit()
```

Aqui, estamos criando o modelo de ANOVA usando a função ols() (Ordinary Least Squares) do módulo statsmodels.formula.api. A fórmula passada para a função ols() descreve a relação entre a variável dependente (Vendas) e a variável independente categórica (Produto). O símbolo '~' indica a relação entre as duas variáveis. A expressão C(Produto) informa que 'Produto' é uma variável categórica (qualitativa).

Em resumo, estamos criando um modelo que relaciona as vendas com o tipo de produto, e o modelo está pronto para ser ajustado aos dados.

```python
# Ajuste do modelo aos dados e cálculo da tabela ANOVA:
anova_table = sm.stats.anova_lm(model, typ=2)
```

Neste passo, ajustamos o modelo criado anteriormente aos dados do DataFrame df usando o método fit(). Isso significa que estamos encontrando os melhores parâmetros do modelo que melhor se ajustam às vendas observadas para cada produto.

A seguir, utilizamos a função anova_lm() do módulo statsmodels.api para calcular a tabela ANOVA. O resultado dessa função é uma tabela que contém as estatísticas relevantes do teste de ANOVA. A opção typ=2 indica o tipo de sumários a serem calculados na ANOVA.

```python
# Exibição do resultado da ANOVA:
print(anova_table)
```

Por fim, estamos imprimindo a tabela ANOVA na tela para visualizar os resultados. A tabela contém informações como soma dos quadrados (sum_sq), graus de liberdade (df), valor F (F) e valor do p-valor (PR(>F)) para a variável categórica 'Produto'.

O valor de F e o p-valor são os principais indicadores que nos ajudarão a decidir se há uma diferença significativa nas vendas dos três produtos. Se o p-valor for menor que um nível de significância escolhido (geralmente 0.05), rejeitamos a hipótese nula e concluímos que há uma diferença estatisticamente significativa nas vendas dos produtos. Caso contrário, não há evidências suficientes para afirmar que as vendas são diferentes entre os produtos.

Nesse bloco de código, primeiro criamos um DataFrame com as vendas e o produto correspondente. Em seguida, realizamos uma ANOVA usando a função ols() do módulo statsmodels.formula.api e a função anova_lm() do módulo statsmodels.api. Finalmente, imprimimos a tabela ANOVA. Se o p-valor na tabela for menor que 0.05, podemos concluir que há uma diferença significativa nas vendas dos três produtos.

O p-valor na tabela ANOVA nos diz a probabilidade de obter os dados observados (ou dados mais extremos) se a hipótese nula for verdadeira. Nesse caso, a hipótese nula é que os três produtos têm as mesmas vendas. Se o p-valor for menor que 0.05, rejeitamos a hipótese nula e concluímos que há uma diferença significativa nas vendas dos três produtos.

<br>

**Memorizando**<br>
Lembre-se da metáfora do show de talentos. ANOVA é como medir o aplauso para cada ato e compará-los para ver se um ato recebe significativamente mais aplausos que os outros.
